```python
def sortSecond(val):
    return val[1] 
```

```python
final_text = doc.text

lst = []
for token in doc:
    lst.append(token)
    
print lst

lst_replace = []

for cluster in doc._.coref_clusters:
    print cluster
    print "Main"
    print cluster.main
    for mention in cluster.mentions:
        print mention
        print mention.start
        print mention.end
        lst_replace.append((mention.start, mention.end, cluster.main.text))

lst_replace.sort(key=sortSecond)
        
print "Finally"
print lst_replace
```
